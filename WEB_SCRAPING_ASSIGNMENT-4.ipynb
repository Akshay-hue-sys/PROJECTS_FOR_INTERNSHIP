```python
#searching for urls for first 3 pages for guitars--
urls=[]

github_urls=gdriver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")  
for i in github_urls:
    urls.append(i.get_attribute('href'))



urls 
```

```python
Repository_title=[]
Repository_description=[]
Number_of_Contributors=[]
Language_used=[]


for ur in urls:

    gdriver.get(ur)

    time.sleep(2)

    try:

        name = gdriver.find_element_by_xpath("//span[@class='author flex-self-stretch']")

        Repository_title.append(name.text)

    except NoSuchElementException:

        Repository_title.append("-")

    try:
        desc_tag=gdriver.find_element_by_xpath("//div[@class='BorderGrid-cell']")
        Repository_description.append(desc_tag.text.replace("\n"," "))
    except NoSuchElementException:
        Repository_description.append('-')

    try:
        nr=gdriver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[3]/div/h2/a")
        Number_of_Contributors.append(nr.text)
    except NoSuchElementException:
        Number_of_Contributors.append('-')  
    
    try:
        l_tags=gdriver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div/ul")
        Language_used.append(l_tags.text.replace("\n",","))
    except NoSuchElementException:
        Language_used.append('-')


    
#creating a dataframe
pd.set_option('display.max_colwidth', None)
df=pd.DataFrame({'Repository_title':Repository_title,'Repository_description':Repository_description,'Number_of_contributors':Number_of_Contributors,
                'Language_used':Language_used})
#printing dataframe
df    
```
